```python
# scaling the data in other rows 
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df3[cols_to_scale] = scaler.fit_transform(df3[cols_to_scale])

df3
```

```python
# converting gender from one form to another 

df3 = df3.replace({'Male': 1, 'Female': 0})
df3
```

```python
# for crating the model we are going to do the train and test split 

x = df3.drop('Churn', axis = 1)
y = df3['Churn']

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5)
 
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
import tensorflow as tf
from tensorflow import keras 

model = keras.Sequential([
    keras.layers.Dense(20,input_shape=(26,),activation='relu'),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(10,activation='relu'),
    keras.layers.Dense(10,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])

model.fit(x_train,y_train,epochs=1000)
```

```python
yp = model.predict(x_test)

yp[:5]
y_pred = []

for elements in yp: 
    if elements >0.5:
        y_pred.append(1)
    else: 
        y_pred.append(0)
    
```

```python
# printing the classification report 

from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf

# Calculate confusion matrix
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_pred)

# Plot
plt.figure(figsize=(10,7))
sns.heatmap(cm.numpy(),  # convert tf.Tensor to numpy
            annot=True, 
            fmt='g',     # 'g' for general number format (not 'f')
            cmap='Reds',  # better color map
            linewidths=0.5,  # add gridlines
            linecolor='black',
            cbar=False)  # remove color bar if not needed
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
def ANN(x_train,y_train,x_test,y_test,loss,weights):
    model = keras.Sequential([
        keras.layers.Dense(26,input_dim=26,activation='relu'),
        keras.layers.Dense(15,activation='relu'),
        keras.layers.Dense(1,activation='sigmoid')
    ])
    
    model.compile(optimizer='adam',loss=loss,metrics=['accuracy'])
    
    if weights ==-1: 
        model.fit(x_train,y_train,epochs=200)
    else:
        model.fit(x_train,y_train,epochs=200,class_weight = weights)
    print(model.evaluate(x_test,y_test))
    
    y_preds = model.predict(x_test)
    y_preds = np.round(y_preds)
    print(classification_report(y_test,y_preds))
    
```

```python
y_preds = ANN(x_train,y_train,x_test,y_test,"binary_crossentropy",-1)
```

```python
# method 1 : undersampling

# class count 
count_class_0 , count_class_1 = df3.Churn.value_counts()

#divide by class 

df_class_0 = df3[df3['Churn'] ==0]
df_class_1 = df3[df3['Churn'] ==1]

print(df_class_0.shape)
print(df_class_1.shape)
```

```python
# we know that class 0 has more number of samples so we will under sample it now 
# to randonly select samples from the df we can you a function called 

df_class_0.sample(2) # it selects two random samples 
```

```python
# taking the numner of sampels as that of the count class 1 

df_class_0_under = df_class_0.sample(count_class_1)
```

```python
#concatenating 2 data frams 0 and 1 

df_test_under = pd.concat([df_class_0_under,df_class_1],axis=0)

df_test_under
```

```python
# creating x and y variable from the data frame 

x= df_test_under.drop('Churn',axis=1)
y= df_test_under['Churn']

```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=15, stratify = y)

# stratify ensures you have balanced samples in yout train test split 
# the y in stratify = y is the y== df_test_under['Churn'] that we habe declared above
```

```python
y_train.value_counts()
```

```python
y_preds = ANN(x_train,y_train,x_test,y_test,'binary_crossentropy',-1)
```

```python
# to duplicate the samples in class 1 
#print(5163-1869)
df_class_1_over = df_class_1.sample(count_class_0,replace=True)

df_test_over = pd.concat([df_class_0,df_class_1_over],axis=0)
```

```python
# creating x and y variable from the data frame 

x= df_test_over.drop('Churn',axis=1)
y= df_test_over['Churn']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=15, stratify = y)
y_preds = ANN(x_train,y_train,x_test,y_test,'binary_crossentropy',-1)
```

```python
# Method 3 : SMOTE

# creating x and y variable from the data frame 

from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification

x= df3.drop('Churn',axis=1)
y= df3['Churn']

smote = SMOTE(sampling_strategy='minority')
x_sm, y_sm = smote.fit_resample(x, y)

y_sm.value_counts()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_sm,y_sm,test_size=0.2,random_state=15, stratify = y_sm)
y_preds = ANN(x_train,y_train,x_test,y_test,'binary_crossentropy',-1)
```
