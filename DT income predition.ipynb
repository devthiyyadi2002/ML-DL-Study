```python
# we can see the datas in the colomns have a ? in the missing areas so inforfer to find the colomns with missing values 
df1 = df[df.workclass == '?']
df1
```

```python
# select all category variables 

df_categorical = df.select_dtypes(include=['object'])

# check whether the ? is in the categorical variables
df_categorical.apply(lambda x: x=="?",axis=0).sum()
```

```python
# dropping the ? values
df = df[df['occupation'] != '?']
df = df[df['workclass'] != '?']
df = df[df['native.country'] != '?']

```

Data Preparation

```python
# we still need to encode the categorical variables into a standard format
# using sklearn we will be using the LabelEncoder() class which comes with sklearn.preprocessing

from sklearn import preprocessing
# encode categorical variables
df_categorical = df.select_dtypes(include=['object'])
df_categorical.head()
```

```python
# apply labbel encoder to df_categorical

le = preprocessing.LabelEncoder()
df_categorical = df_categorical.apply(le.fit_transform)
df_categorical.head()

# it assigns the object values to a specific number 
```

```python
# concat df_categorical with origiinal df 

df = df.drop(df_categorical.columns, axis=1)
df = pd.concat([df,df_categorical], axis=1)
df.head()

df
```

```python
df.info()
```

```python
# convert target variable invome to categorical 

df['income'] = df['income'].astype('category')
```

Let's build the model

```python
# first lets build a DT with default hyper parameters then we will use cross validation and tune them 
from sklearn.model_selection import train_test_split

# putting features into x variable 

x=df.drop('income',axis=1)
# putting responce variable into y variable
y=df['income']
```

```python
# splitting the data into train and test sets

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=99)
x_train.head()
```

```python
# importing DT classifier 
from sklearn.tree import DecisionTreeClassifier

# fitting the decision tree classifier with default hyper parameters
dt_default = DecisionTreeClassifier(max_depth =5)
dt_default.fit(x_train,y_train)
```

```python
# lets chheck the evaluation metrics of our default model 

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# making the predictions 
y_pred_default = dt_default.predict(x_test)

```

```python
# printing the confusion matrix
print(confusion_matrix(y_test,y_pred_default))
```
